```python
import os, sys
from tqdm import tqdm

# graphicalModel.py saves to ../../summary_outputs/Graphical_Model/summary_{fraction}
# Pre-create those directories before running
for frac in [0.3, 0.5, 0.7]:
    os.makedirs(
        f"/content/legaldocsvalution/summary_outputs/Graphical_Model/summary_{frac}",
        exist_ok=True
    )

%cd /content/legaldocsvalution/summarization_models/Graphical_Model
import graphicalModel

for summary_fraction in [0.3, 0.5, 0.7]:
    for i in tqdm(range(93), desc=f"fraction={summary_fraction}"):
        graphicalModel.get_summary(f"data/{i+1}.txt", i+1, summary_fraction)

print("Done! Outputs in /content/legaldocsvalution/summary_outputs/Graphical_Model/")
```
