```python
# Clone repo and prepare directories
!git clone https://github.com/Aditya-dom/legaldocsvalution.git /content/legaldocsvalution -q
import os
os.makedirs("/content/berts", exist_ok=True)
```

```python
from summarizer import Summarizer
from tqdm import tqdm

model = Summarizer()

def bert(no):
    path = f"/content/legaldocsvalution/dataset/indian_data/ind_text/{no}.txt"
    with open(path, "r", errors="ignore") as f:
        text = f.read()
    result = model(text, max_length=500)
    summary = "".join(result)
    with open(f"/content/berts/summary_bert_{no}.txt", "w") as out:
        out.write(summary)

for no in tqdm(range(1, 94)):
    bert(no)

print("Done! Summaries saved to /content/berts/")
```
