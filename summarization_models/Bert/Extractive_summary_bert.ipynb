```python
from summarizer import Summarizer

def bert(no):
  f=open('bert/'+str(no)+'.txt','r').read()
  model = Summarizer()
  result = model(f,max_length=500)
  summary = "".join(result)
  with open('berts/summary_bert_'+str(no)+'.txt','w') as f1:
    f1.writelines(summary)

no=1
for i in range(1,94):
  bert(no)
  no+=1
```
